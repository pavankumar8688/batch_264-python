```python
selected_pokemon = pokemon.loc[[0, 1, 2], ['name_x', 'attack', 'defense', 'speed', 'hp', 'sp_attack', 'sp_defense']]

#Normalize the stats to be between 0 and 1
normalized_stats = selected_pokemon[['attack', 'defense', 'speed', 'hp', 'sp_attack', 'sp_defense']].div(255)

#Calculate the number of variables
num_vars = len(normalized_stats.columns)

#Calculate the angle of each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

#Close the plot
values = normalized_stats.loc[0].values.tolist()
values += values[:1]
angles += angles[:1]

#Plot the spider plot for the first Pokémon
plt.figure(figsize=(6, 7), facecolor='dimgray')
plt.polar(angles, values, marker='o', linestyle='solid', linewidth=2, label=selected_pokemon.loc[0, 'name_x'])

#Close the plot for the second Pokémon
values = normalized_stats.loc[1].values.tolist()
values += values[:1]
plt.polar(angles, values, marker='o', linestyle='solid', linewidth=2, label=selected_pokemon.loc[1, 'name_x'])

#Close the plot for the third Pokémon
values = normalized_stats.loc[2].values.tolist()
values += values[:1]
plt.polar(angles, values, marker='o', linestyle='solid', linewidth=2, label=selected_pokemon.loc[2, 'name_x'])

#Label each axis
plt.xticks(angles[:-1], normalized_stats.columns)

#Show the plot
plt.title('Spider Plot of Pokémon Stats')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Assuming 'your_dataset' is your dataset with the given column names
# Replace 'your_dataset' with the actual variable name of your dataset

# Choose the variables for the spider plot
variables = ['attack', 'defense', 'sp_attack', 'sp_defense', 'speed']

# Select a subset of your dataset for these variables
subset = pokemon[variables]

# Normalize the data to a [0, 1] range for better comparison
normalized_data = (subset - subset.min()) / (subset.max() - subset.min())

# Number of variables
num_vars = len(variables)

# Compute angle of each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Repeat the first angle to close the circle
values = normalized_data.iloc[0].tolist()
values += values[:1]
angles += angles[:1]

# Plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
ax.fill(angles, values, color='skyblue', alpha=0.25)

# Plot each individual Pokémon's values
for i in range(len(normalized_data)):
    values = normalized_data.iloc[i].tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=2, linestyle='solid')

# Add labels for each variable
ax.set_thetagrids(np.degrees(angles[:-1]), variables)

# Add a legend
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show the plot
plt.title('Spider Plot of Pokemon Attributes')
plt.show()
```
